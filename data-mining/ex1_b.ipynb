## **Utilize o dataset Iris para a computar a matriz de distância entre todos os os objetos, considerando todas as features. Utilize a distância euclideana para essa tarefa.**

```python
import csv
import numpy as np
import math

data = []
with open('datasets/Iris.csv', 'r') as f:
    lines = csv.reader(f)
    next(lines)
    for line in lines:
        data.append([float(row) for row in line[1:5]])

data = np.array(data)
dist = lambda p1, p2: math.sqrt(((p1-p2)**2).sum())
dm = np.asarray([[dist(p1, p2) for p2 in data] for p1 in data])
print(dm)
```

```python
def enumerize(col):
    values_set = set(col)
    i = 0
    values_dict = {}
    for v in values_set:
        values_dict[v] = i
        i += 1
    return col.map(values_dict).astype('int8')

df0 = df.copy()
df0 = df0.drop(columns=['survived'])
df0['sex'] = enumerize(df0['sex'])
df0['class'] = enumerize(df0['class'])
df0['alive'] = enumerize(df0['alive'])
df0['who'] = enumerize(df0['who'])
df0.dtypes
```
