```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori

cset  =   [['tree','cup', 'paper'],
                         ['book','tomato','pen'],
                         ['pen', 'book', 'tree'],
                         ['tomato','pen', 'cup'],
                         ['tree','paper', 'book'],
                         ['pen','book']]

te = TransactionEncoder()
encoded = te.fit(cset).transform(cset)
print(encoded)

df = pd.DataFrame(encoded, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.33, use_colnames=True)
frequent_itemsets
```

```python
cset  =   [[1, 3, 5, 7, 8],
           [3, 4, 5, 7],
           [2, 3, 5, 6],
           [2, 4, 5, 8, 9],
           [2, 3, 5, 11],
           [1, 2, 3, 7, 9]]

te = TransactionEncoder()
encoded = te.fit(cset).transform(cset)
print(encoded)

df = pd.DataFrame(encoded, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.73, use_colnames=True)
frequent_itemsets
```

```python
dfMercado = pd.read_csv('datasets/supermercado.csv', header=None, sep=';')
items = [s[0].split(',') for s in dfMercado.values]
items = [list(set(i)) for i in items]
items
```

```python
from mlxtend.frequent_patterns import association_rules

te = TransactionEncoder()
encoded = te.fit(items).transform(items)
print(encoded)

df = pd.DataFrame(encoded, columns=te.columns_)

frequent_itemsets = apriori(df, min_support=0.001, use_colnames=True)
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.25)
rules
```

```python
beer_rules = [True if 'bottled beer' in i else False for i in rules['consequents']]
rules[beer_rules]
```
